```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import pandas as pd
import math
```

**Reading the data**

```python
df = pd.read_csv('data/first_data.txt',header=None,names=['City Population','Restaurant_Profit'])
df
```

**Modifying the columns name for uniformaty**

```python
df = df.rename(columns={i:i.lower().replace(' ','_') for i in df.columns})
df
```

**Column city_population represent the population of city, whose actual value is the value times 10,000 and the column average_profit represent the profit of restaurant in each city in the units of $10,000**

**Studying the data**

```python
X_train = np.array(df['city_population'])
m=X_train.shape[0]
print(f'The number of training examples is {m}')
Y_train = np.array(df['restaurant_profit'])
print(f'The profit of a restaurant in each city is {Y_train}')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_train,Y_train,marker='x',c='b')
plt.title('Profit vs Population per city')
plt.xlabel('Population of city in 10,000d')
plt.ylabel('Profit of restaurant in $10,000')
plt.show()
```

**Cost Function**

```python
def cost_function(w,x,b,y):
    m= x.shape[0]
    total_cost = 0
    for i in range(m):
        y_pred = np.dot(x[i],w) + b
        error = y_pred - y[i]
        sqaured_error = np.square(error)
        total_cost+=sqaured_error
    return total_cost / (2*m)    
w_init = 2
b_init = 1
cost = cost_function(w_init,X_train,b_init,Y_train)
print(f'The initial cost is: {cost:.3f}')
    
        
```

**Computing the partial derivative**

```python
def gradient(w,x,b,y):
    m = x.shape[0]  #number of training examples
    dj_dw = 0  #partial derivative with respect to weight
    dj_db = 0  #partial derivative with respect to bias
    for i in range(m):
        y_pred = np.dot(x[i],w) + b
        error = y_pred - y[i]
        dj_db+=error
        dj_dw+= error * x[i]
    return dj_dw/m,dj_db/m
dj_dw,dj_db =  gradient(w_init,X_train,b_init,Y_train)   
print(f'The partial derivatives are: {dj_dw,dj_db} derivative of cost function with respect to weight and bias respectively')        
```

**Training the model using gradient descent**

```python
def gradient_descent(w_init,x,b_init,y,gradient,cost_function,alpha,iterations):
    m = x.shape[0]
    w = copy.deepcopy(w_init)
    b = b_init
    cost_history = []
    iterations_history = []
    for i in range(iterations):
        cost = cost_function(w,x,b,y)
        dj_dw,dj_db = gradient(w,x,b,y)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        if i % math.ceil(iterations/10) ==0:
            cost_history.append(cost)
            iterations_history.append(i)
    return w,b,cost_history, iterations_history
initial_w = 0.
initial_b = 0.
# some gradient descent settings
iterations = 1500
alpha = 0.01    
final_w,final_b,cost_history, iterations_history = gradient_descent(initial_w,X_train,initial_b,Y_train,gradient,cost_function,alpha,iterations)        
print(f'The parameters of the model are weight:{final_w} and bias:{final_b} respectively')        
```

**plotting the learning curve**

```python
plt.plot(iterations_history,cost_history)
plt.xlabel('Number of iterations')
plt.ylabel('Loss Function')
plt.show()
```

**As we can observe from the learning curve that the loss function is decreasing overtime and is converging after about 700 iterations approximately**
